# Marching cube playground

## Using Skicit-image

### Init

```python
import numpy as np

from skimage import measure

def sphere_sdf(x, y, z, radius=10):
    return np.sqrt(x**2 + y**2 + z**2) - radius

size = 25
coords = np.linspace(-15, 15, size)
X, Y, Z = np.meshgrid(coords, coords, coords)

sphere = sphere_sdf(X, Y, Z, radius=10)

# Use marching cubes to obtain the surface mesh of the sphere
verts, faces, normals, values = measure.marching_cubes(sphere, 0, step_size=1, spacing=(1, 1, 1))
```

### Plotly visualisation

```python
import plotly.graph_objects as go

vertex_colors = ((normals + 1) / 2 * 255).astype(int)
rgb_colors = ['rgb({},{},{})'.format(vertex_colors[i, 0], vertex_colors[i, 1], vertex_colors[i, 2]) 
              for i in range(len(vertex_colors))]


fig = go.Figure(data=[go.Mesh3d(
    x=verts[:, 0],
    y=verts[:, 1],
    z=verts[:, 2],
    i=faces[:, 0],
    j=faces[:, 1],
    k=faces[:, 2],
    vertexcolor=rgb_colors,
    opacity=0.9,
    flatshading=False,
    showlegend=True,
    name='Sphere Surface'
)])


fig.update_layout(
    title='3D Sphere Mesh Visualization',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
        aspectmode='data',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    width=900,
    height=900,
    showlegend=True
)

fig.show()
```

### Open3d Visualisation

```python
import open3d as o3d

# Create the mesh with vertices and triangles
mesh = o3d.geometry.TriangleMesh(
    vertices=o3d.utility.Vector3dVector(verts),
    triangles=o3d.utility.Vector3iVector(faces)
)

o3d.visualization.draw([mesh], show_ui=True)
```
